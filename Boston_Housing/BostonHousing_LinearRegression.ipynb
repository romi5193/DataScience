### Plot distribution of Median Home prices

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.distplot(boston['medv'], bins=20)
plt.show()
```

### Create correlation matrix to identify features that strongly correlate with the Median Home prices

```python
correlation_matrix = boston.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```

```python
all_features = boston[['crim','zn','indus', 'chas','nox', 'rm', 'age', 'dis', 'rad','tax','ptratio', 'black', 'lstat']]
features = boston[['rm', 'lstat','ptratio','indus']]
target = boston[['medv']]
```

```python
plt.figure(figsize=(20,10))
for i, col in enumerate(features):
    plt.subplot(2, 2 , i+1)
    x = boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('medv')
plt.show()
```

# Concatenate corelated data into new dataframe

```python
X = features
Y = boston['medv']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 0.8, random_state=6)
```

```python
reg_model = LinearRegression()
reg_model.fit(x_train, y_train)
reg_model.score(x_train, y_train)
```

```python
y_predict=reg_model.predict(x_test)
reg_model.score(x_test, y_test)
```
